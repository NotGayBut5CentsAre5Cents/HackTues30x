```python
import pickle

with open('./articles.p', 'rb') as fp:
    heds, contents = pickle.load(fp)
```

```python
from collections import Counter
from itertools import chain
```

```python
import re
import nltk
all_text = " ".join(heds + contents).replace(".", ". ")
tokenized_words = nltk.word_tokenize(all_text)
vocabcount = Counter(word for word in tokenized_words)
vocab = map(lambda x: x[0], sorted(vocabcount.items(), key=lambda x: -x[1]))
```

```python
len(vocabcount)
```

```python
glove_vec_size = 100
fname = 'glove.6B.%d.txt'%glove_vec_size
import os
start = 2
```

```python
word2idx = dict((word, i+start) for i,word in enumerate(vocab))
```

```python
word2idx['<empty>'] = 0
word2idx['<eos>'] = 1
```

```python
idx2word = dict((idx,word) for word,idx in word2idx.items())
```

```python
import os
glove_path = r'./data/glove.6B/glove.6B.100d.txt'
```

```python
glove_word_count = !wc -l {glove_path}
```

```python
glove_word_count = int(glove_word_count[0].split(" ")[0])
```

```python
import numpy as np
glove_index_dict = {}
glove_embedding_weights = np.empty((glove_word_count, 100))
with open(glove_path, 'r') as fp:
    i = 0
    for l in fp:
        l = l.strip().split() #split vector values
        word = l[0]
        glove_index_dict[word] = i
        slice = map(float,l[1:])
        glove_embedding_weights[i,:] = list(slice)
        i += 1
```

```python
glove_embedding_weights
```

```python
for w,i in glove_index_dict.items():
    l = w.lower()
    if w not in glove_index_dict:
        glove_index_dict[l] = i
        del glove_index_dict[w]
```

```python
np.random.seed(42)
shape = (glove_word_count, glove_vec_size)
scale = glove_embedding_weights.std()
embedding = np.random.uniform(low=-scale, high=scale, size=shape)

c = 0
for i in range(len(vocabcount)):
    word = idx2word[i]
    glove_idx = glove_index_dict.get(word) or glove_index_dict.get(word.lower())
    if glove_idx is not None:
        embedding[i,:] = glove_embedding_weights[glove_idx,:]
        c+=1
    else:
        print(word)
c
```
