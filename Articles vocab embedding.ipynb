```python
import pickle

with open('./articles.p', 'rb') as fp:
    heds, contents = pickle.load(fp)
    
from collections import Counter
from itertools import chain
```

```python
import re
import nltk
all_text = " ".join(heds + contents).replace(".", " . ").replace("'", " ' ")
tokenized_words = nltk.word_tokenize(all_text)
vocabcount = Counter(word.strip() for word in tokenized_words)
vocab = list(map(lambda x: x[0], sorted(vocabcount.items(), key=lambda x: -x[1])))
```

```python
glove_vec_size = 100
glove_path = '/home/martin/HackTues30x/data/glove.6B/glove.6B.100d.txt'
import os
start = 2
```

```python
import numpy as np
glove_index_dict = {}
glove_embedding_weights = np.empty((400000, glove_vec_size))
c = 0
with open(glove_path, 'r') as fp:
    i = 0
    for l in fp:
        l = l.strip().split() #split vector values
        word = l[0]
        c+=1
        glove_index_dict[word] = i
        slice = map(float,l[1:])
        glove_embedding_weights[i,:] = list(slice)
        i += 1
```

```python
our_words_in_glove = {}
i = 3
for word in vocabcount:
    if(word in glove_index_dict or word.lower() in glove_index_dict):
        if(not word.lower() in our_words_in_glove):
            our_words_in_glove[word.lower()] = i
            i+=1
    else:
        #print(word)
        pass
```

```python
our_embedding = np.empty((len(our_words_in_glove) + 3, glove_vec_size))
our_index = {}
for word in our_words_in_glove:
    idx = our_words_in_glove[word]
    glove_idx = glove_index_dict[word]
    our_index[word] = idx
    our_embedding[idx] = glove_embedding_weights[glove_idx]
    
```
